```python

import copy


def my_input():

    n,m = map(int, input().split())
    people = 0
    #people
    str = input()
    if str != '0':
        people = list(map(int, str.split()))
        num_people = people.pop(0)
    else:
        people = -1
        num_people = -1


    party = []
    for i in range(m):
        part = list(map(int, input().split()))
        party.append(part)

    return n,m,num_people,people,party


n,m,num_people,people,party = my_input()
visited = [0 for _ in range(m)]
if num_people == -1:
    print(m)
    exit()

trueList = []

for i in range(m):
    party[i].pop(0)
    
trueList = copy.deepcopy(people)
is_updated = 0

def add():
    #각 파티에 대해 반복
    for i in range(m):
        global is_updated
        #진실을 아는 사람의 리스트 반복
        for a in trueList:
            #진실을 아는사람이 파티i에 있으면,
            if a in party[i]:
                #파티i에 있는 사람 모두를 추가(중복 빼고)
                for j in range(len(party[i])):
                    if(party[i][j] not in trueList):
                        trueList.append(party[i][j])
                        is_updated = 1
                break
        
while(True):
    add()
    if(is_updated == 0):
        break
    else:
        is_updated = 0
for i in range(len(party)):

    for a in trueList:

        if a in party[i]:
            visited[i] = 1
count = 0
for i in visited:
    if i == 0:
        count += 1
print(count)
```
